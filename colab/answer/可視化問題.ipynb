```python
#使うライブラリ
#pip install japanize-matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()
```

## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
path = '../quizzes/matplotlib_quiz/hist_data.csv'
data = np.loadtxt(path,dtype=None,skiprows=1,delimiter = ",",encoding="utf-8")
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(data[:,0], bins=30)
ax.hist(data[:,1], bins=30,color=(1, 0, 0, 0.3))
ax.set_title('aデータとbデータの温度に関する度数分布')
ax.set_xlabel('温度(℃)')
ax.set_ylabel('度数')
ax.legend(['aデータ','bデータ'])
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
path ='../quizzes/matplotlib_quiz/scatter_data.csv'
data = np.loadtxt(path,skiprows=1,delimiter=",",encoding="utf-8")
fig = plt.figure()

ax = fig.add_subplot(1,1,1)
ax.scatter(data[:,0][data[:,2]==0],data[:,1][data[:,2]==0],color='red')
ax.scatter(data[:,0][data[:,2]==1],data[:,1][data[:,2]==1],color='blue')
ax.scatter(data[:,0][data[:,2]==2],data[:,1][data[:,2]==2],color='green')
ax.set_title('クラス0~2のlengthとwidthに関する散布図')
ax.set_xlabel('length(cm)')
ax.set_ylabel('width(cm)')
ax.legend(['class0','class1','class2'])
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

問3(1).png

問3(2).png

```python
path ='../quizzes/matplotlib_quiz/line_data.csv'
df = pd.read_csv(path)

num = range(len(df['件数']))
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.plot(num,df['件数'],color='b',marker='o')
ax1.plot(num,df['件数'],color='b',ls='-')
label = ax1.set_xticklabels(df['期間'])
plt.setp(label,rotation=45)
ax1.set_title('期間ごとの件数の推移')
ax1.set_xlabel('期間')
ax1.set_ylabel('件数')

fig = plt.figure()
ax2 = fig.add_subplot(1,1,1)
ax2.bar(num,df['費用'],color='b')
label = ax2.set_xticklabels(df['期間'])
plt.setp(label,rotation=45)
ax2.set_title('期間ごとの費用の推移')
ax2.set_xlabel('期間')
ax2.set_ylabel('費用(円)')
```
